## 8. Plotting with Matplotlib Library

```python
import numpy as np
import matplotlib.pyplot as pl
#additionally use %matplotlib inline, if you are using ipyhton notebook.

f = 1
t = np.linspace(0,10,500)
y = np.sin(2*np.pi*f*t)*np.exp(-0.5*t)
pl.plot(t,y)
pl.title('Damped Sine Wave')
pl.xlabel('t (s)')
pl.ylabel('y')
pl.grid()
pl.show()
```
